This Python project analyzes the Hotel Bookings dataset to understand customer behavior, booking trends, and revenue performance in the hospitality sector. Visualizations like bar charts and heatmaps help present findings clearly. Overall, the project demonstrates how Python can be used to clean, analyze, and visualize real-world data to support data-driven decision-making in the hotel industry.

## Data Wrangling

```python
# Write your code to make your data analysis ready
# Drop duplicate value
df1.drop_duplicates()
```

```python
null_columns = ['company', 'agent', 'country', 'children']
for columns in null_columns:
    df1[columns].fillna(0, inplace = True)
```

```python
df1['country'].fillna('others', inplace = True)
```

```python
df1.isnull().sum().sort_values(ascending = True)
```

```python
# Adding total staying day in hotel.
df1['Total_staying_day'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']
df1
```

```python
# Adding total number of people
df1['Total_people'] = df1['babies'] + df1['children'] + df1['adults']
df1
```

```python
df1[df1['children'] + df1['babies'] + df1['adults'] == 0].shape
```

```python
df1.drop(df1[df1['children'] + df1['babies'] + df1['adults'] == 0].index, inplace = True)
df1
```

## 4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

## Chart 1

```python
# 1. Which type of hotel is mostly prefer by guests.
hotel_value_count = df1['hotel'].value_counts()
print(hotel_value_count)
```

```python
hotel_value_count.plot.pie(
    explode=[0.05, 0.05],   # length must match categories
    autopct="%1.1f%%",
    shadow=True,
    figsize=(8,6),
    fontsize=12
)

plt.title("Hotel Preference")
plt.ylabel("")   # remove default y-label
plt.show()
```

### 1️⃣ Why this chart?
-> Pie chart is used to show percentage distribution between hotel types clearly.

### 2️⃣ Insights found?
 -> One hotel type has higher bookings, showing stronger customer preference.

### 3️⃣ Business impact? 
 -> Yes, it helps in better pricing, marketing, and resource planning. Lower booking share may indicate weaker demand and potential revenue loss.

## Chart 2

```python
# Which agent make the highest booking??
agent_count = df1['agent'].value_counts().reset_index().rename(columns={'agent':'agent_id' , 'count':'booking_count'})[:10]
print(agent_count)
```

```python
plt.figure(figsize=(10,6))

plt.bar(agent_count['agent_id'].astype(str),
        agent_count['booking_count'], color = 'red')

plt.xlabel("Agent ID")
plt.ylabel("Number of Bookings")
plt.title("Top 10 Agents by Booking Count")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = agent_count['agent_id'], 
            y = agent_count ['booking_count'], 
            order = agent_count['agent_id'])
plt.xlabel("Agent number", fontsize = 15)
plt.ylabel("Number of Bookings", fontsize = 15)
plt.title("Most booking by agent", fontsize = 20)
plt.xticks(rotation=45)

plt.show()
```

## 1️⃣. Why did you pick the specific chart?
 -> A bar chart was chosen because it is best for comparing booking counts across different agent IDs (categorical comparison).

## 2️⃣. What is/are the insight(s) found from the chart?
 -> The chart shows that a few agents contribute significantly higher bookings, indicating unequal performance distribution among agents.

## 3️⃣. Will the gained insights help creating a positive business impact? Are there any insights that lead to negative growth?
 -> Yes, identifying top and low-performing agents helps optimize incentives and training; low-performing agents may indicate potential revenue loss if not improved.

## Chart 6

```python
# 6. In which month most of the booking happened.
month_order = [
    'January', 'February', 'March', 'April',
    'May', 'June', 'July', 'August',
    'September', 'October', 'November', 'December'
]
Most_bookings = df1['arrival_date_month'].value_counts().reindex(month_order)
Most_bookings
```

## Chart 11

```python
# Select only numerical columns
numeric_df = df1.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, 
            cmap='coolwarm',
            linewidths=0.5)

plt.title("Correlation Heatmap", fontsize=18, fontweight='bold')
plt.show()
```

## 1️⃣ Why this chart?
 -> Heatmap is used to visualize the correlation between numerical variables in the dataset.

## 2️⃣ Insights found?
 -> Some variables show strong positive or negative correlation, indicating relationships such as higher lead time potentially affecting cancellation rates.

## 3️⃣ Business impact?
 -> Yes, understanding correlations helps in identifying key factors influencing cancellations, revenue (ADR), and booking behavior.
Strong negative correlations may indicate risk factors affecting business performance.
